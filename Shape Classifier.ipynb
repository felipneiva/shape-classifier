Introducing "Shape Classifier" - a website that can easily classify any image as one of three iconic shapes - Triangle, Circle, or Square - using deep learning technology.

Simply upload an image and let the Shape Classifier do the rest. Within seconds, you'll receive a detailed analysis of the image.

```python
import pathlib
import fastai.vision.all as vision
import ipywidgets as widgets
from IPython.display import display
```

```python
path = pathlib.Path()
learn_inf = vision.load_learner(path/'export.pkl')
```

```python
uploader = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
btn_run = widgets.Button(description='Classify')
```

```python
def classify(change):
    img = vision.PILImage.create(uploader.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred, idx, probs = learn_inf.predict(img)
    if isinstance(probs, vision.TensorImage):
        probs = probs.numpy()
    prob = probs[idx.item()]
    lbl_pred.value = f'Prediction: {pred}; Probability: {prob:.04f}'
```

```python
btn_run.on_click(classify)
```

```python
widgets.VBox([
    widgets.Label('Select your Shape!'),
    uploader, btn_run, out_pl, lbl_pred
])
```
